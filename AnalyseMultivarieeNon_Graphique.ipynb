  ## **Introduction à l'exploration multivariée non-graphique**
 L'exploration multivariée non-graphique s'intéresse à l'analyse de deux ou plusieurs variables à la fois. Elle étudie les relations existantes entre ces variables et utilise différentes  techniques statistiques pour illustrer ces relations tel que les tableaux de contingence et les matrices de corrélation.


  ## **Techniques d'exploration multivariée non-graphique**


 ### **1.   Tableau de contingence**

Le tableau de contingence permet d'illustrer les fréquences représentantes des modalités croisées de deux attributs x et y. Le tableau de contingence est utilisé seulement dans le cas de variables qualitatives.

### **2. Matrice de corrélation**

La matrice de corrélation permet d'illustrer les relations de dépendance entre deux attributs x et y. La matrice de corrélation est utilisée seulement dans le cas de variables quantitatives. On distingue deux types de corrélations:

* **Corrélation positive:** dans ce cas les deux attributs ou variables varient dans le même sens.
  
image.png

* **Corrélation négative:** dans ce cas la première variable varie dans le sens inverse de la deuxième variable.
