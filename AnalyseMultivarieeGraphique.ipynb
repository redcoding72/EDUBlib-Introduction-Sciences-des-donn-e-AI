  ## **Introduction à l'exploration multivariée graphique**
 L'exploration multivariée graphique, appelée aussi visualisation multivariée, s'intéresse à l'analyse de deux ou plusieurs variable à la fois. Elle étudie les relations existantes entre ces variables à l'aide des graphiques tels que: le nuage de points, le diagramme à barres, la carte thermique, etc.


  ## **Exploration multivariée graphique**
Différentes étapes sont nécessaires pour la visualisation multivariée:

* Importer les modules Python necessaires
* Charger les données et les sauvegarder dans un *DataFrame*
* Visualiser les données à l'aide d'un graphique.

Commencez à exécuter le code suivant pour importer le module *Seaborn*, charger les données dans un *DataFrame* et avoir un aperçue sur les données du cancer de sein.

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
import seaborn as sns
cancerData = load_breast_cancer()
cancer = pd.DataFrame(cancerData.data, columns=cancerData.feature_names)
cancer.head()
```

 ### **1. Nuage de points**
Le nuage de points est une technique de visualisation multivariée permettant d'illustrer les relations entres deux variables quantitatives. Il permet aussi de détecter les valeurs abérrantes.
Afin d'illustrer un nuage de points en Python, vous pouvez utiliser la fonction *scatterplot* de *seaborn*. Les deux variables à analyser (*"mean perimeter"* et *"mean area"*) doivent être passées en paramètres à la fonction. N'oublier pas d'exécuter le code suivant pour visualiser le nuage de points.


```python
sns.scatterplot(data=cancer,x="mean perimeter",y="mean area")
```

### **2. Carte thermique**
La carte thermique est aussi une technique de visualisation multivariée pusiqu'elle permet d'illustrer entre autres les relations entre plusieurs variables à la fois.
Elle permet de représenter la matrice de corrélation à l'aide de dégradation de couleurs.
En *Python*, on peut utiliser la fonction *heatmap* pour dessiner une carte thermique.



```python
sns.heatmap(data=cancer.corr().round(2), cmap='coolwarm')
```
